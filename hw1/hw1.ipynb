```python
import numpy as np
import matplotlib.pyplot as plt
```

a) Равномерное распределение $U[0,\theta] = \theta\cdot U[0,1]$
$$\theta^{*} = \left((k+1)\cdot\overline{X^k}\right)^{1/k}$$

```python
MAX_k = 100     # максимальная степень момента
MAX_s = 1000    # размер выборки
MAX_ex = 1000   # количество экспериментов
theta_real = 1  # заданный параметр theta
s = []          
theta = []
for i in range(MAX_ex):
    s = np.random.uniform(0, 1, MAX_s) * theta_real
    S = [1] * MAX_s
    th = []
    for k in range(1, MAX_k+1):
        S *= s
        th.append((np.mean(S)*(k+1))**(1/k))
    theta.append(th)
theta=np.array(theta)
theta_std=[]
for i in range(MAX_k):
    theta_std.append(np.sqrt(((theta[:, i]-theta_real)**2).mean()))
plt.plot(np.linspace(1,MAX_k,MAX_k), theta_std)
plt.xlabel('k')
plt.ylabel(r'std $\theta^{*}$')
plt.show()
```

Для лучшей оценки параметра $\theta$ равномерного распределения стоит брать момент большой степени.

b) Равномерное распределение $\text{Exp}(\theta) = \theta\cdot \text{Exp}(1)$
$$\theta^{*} = \left(\dfrac{\overline{X^k}}{k!}\right)^{1/k}$$

```python
MAX_k = 100
MAX_s = 1000
MAX_ex = 1000
theta_real = 1
s = []
theta = []
for i in range(MAX_ex):
    s = np.random.exponential(1, MAX_s) * theta_real
    S = [1] * MAX_s
    th = []
    k_fact = 1
    for k in range(1, MAX_k+1):
        S *= s
        k_fact *= k
        th.append((np.mean(S)/k_fact)**(1/k))
    theta.append(th)
theta=np.array(theta)
theta_std=[]
for i in range(MAX_k):
    theta_std.append(np.sqrt(((theta[:, i]-theta_real)**2).mean()))
plt.plot(np.linspace(1,MAX_k,MAX_k), theta_std)
plt.xlabel('k')
plt.ylabel(r'std $\theta^{*}$')
plt.show()
```

Для лучшей оценки параметра $\theta$ экспоненциального распределения стоит брать момент малой степени.
