```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MAX_k = 100
MAX_s = 1000
MAX_ex = 1000
theta_real = 1
s = []
theta = []
for i in range(MAX_ex):
    np.random.seed(i)
    s = np.random.uniform(0,1,MAX_s) * theta_real #+ np.random.normal(0, 0.04, MAX_s)
    S = [1] * MAX_s
    th = []
    for k in range(1,MAX_k+1):
        S *= s
        th.append((np.mean(S)*(k+1))**(1/k))
    theta.append(th)
theta=np.array(theta)
theta_std=[]
for i in range(MAX_k):
#     theta_std.append(theta[:, i].std())
    theta_std.append(np.sqrt(((theta[:, i]-theta_real)**2).mean()))
plt.plot(np.linspace(1,MAX_k,MAX_k), theta_std)
plt.xlabel('k')
plt.ylabel('std')
plt.show()
```

```python
MAX_k = 100
MAX_s = 1000
MAX_ex = 1000
theta_real = 1
s = []
theta = []
for i in range(MAX_ex):
    s = np.random.exponential(1, MAX_s)*theta_real
    S = [1] * MAX_s
    th = []
    k_fact = 1
    for k in range(1,MAX_k+1):
        S *= s
        k_fact *= k
        th.append((np.mean(S)/k_fact)**(1/k))
    theta.append(th)
theta=np.array(theta)
theta_std=[]
for i in range(MAX_k):
#     theta_std.append(theta[:, i].std())
    theta_std.append(np.sqrt(((theta[:, i]-theta_real)**2).mean()))
plt.plot(np.linspace(1,MAX_k,MAX_k), theta_std)
plt.xlabel('k')
plt.ylabel('std')
plt.show()
```
